# Abstract

The emotional states of college students always fluctuate throughout the academic year, especially during midterms and finals season. Typically, students experiencing extreme emotions of stress and anxiety (negative emotions) turn to social media– especially Twitter– to air out their grievances. Twitter serves as a space where students can freely express their emotions and sentiments, providing a large source of information to get data from. Our study specifically investigates how the emotional sentiment of undergraduate college students changes over the time frame of April-June (comparing sentiment from all of the spring quarter to finals week of the spring quarter). Our study proves to be very important because it determines if exam-related stress has a significant impact on positive or negative online mental health discussions with college students. 

To explore our question, we conducted sentiment analysis using 1.6 million tweets from the Sentiment140 dataset. This dataset was helpful and usable because the tweets in the dataset were labeled as positive, negative, or neutral sentiment. Through filtering the tweets, we were able to see how talk related to finals season changed over time. Our major findings show that negative sentiment increases during midterms and finals season, which supports our hypothesis that depression and anxiety-related discussions increase during these academic periods. As a whole, we can conclude that emotional patterns do shift throughout the academic year– increasing negative sentiment and that mental health discussions intensify during high-stress periods of midterms and finals.

## Data Cleaning
In our dataset, we decided to remove the following columns “user” and “flag” for privacy and relevancy concerns. To protect user privacy, we removed the “user” column that contained the email addresses of the individuals who tweeted their sentiments, ensuring their anonymity. We also removed the “flag” column as it contained the response “NO_QUERY”, which lacked meaningful context and relevance to our research question. Since these columns serve no purpose in our analysis, removing them helped make the dataset efficient, maintain focus towards our research question, and improve data organization. Lastly, we renamed target to sentiment for clarity.


```python
## Organizing text column to keep only college related tweets

# Define the words to filter out college-related tweets
# Require "college" or "university"
college_words = ["college", "university"]

# Academic-related words
academic_words = ["exam", "midterm", "finals", "quiz", "test", "GPA", "grade", "homework", "assignment"]

# Context words related to academic stress
context_words = [
    "cram", "late night", "all-nighter", "procrastination", "burnout", "stress", "anxiety", "overwhelmed",
    "depression", "mental health", "panic attack", "breakdown", "crying", "pressure", "perfectionism",
    "self-care", "counseling", "therapy", "sleep deprived", "tired", "caffeine", "coffee", "energy drink",
    "overworked", "underprepared", "passing", "failing", "academic probation", "cheating", "curve",
    "professor", "syllabus", "study session", "group project", "brain fog", "time management",
    "self-doubt", "motivation", "imposter syndrome", "frustrated", "struggling", "coping", "burnout recovery"
]

# Regex patterns
college_pattern = r'\b(?:' + '|'.join(college_words) + r')\b'
academic_pattern = r'\b(?:' + '|'.join(academic_words) + r')\b'
context_pattern = r'\b(?:' + '|'.join(context_words) + r')\b'

# Step 1: Require "college" or "university"
tweets_college = tweets_140[tweets_140['text'].str.contains(college_pattern, case=False, na=False, regex=True)]

# Step 2: Keep only tweets that also mention an academic word OR a context word
tweets_college = tweets_college[
    tweets_college['text'].str.contains(academic_pattern, case=False, na=False, regex=True) |
    tweets_college['text'].str.contains(context_pattern, case=False, na=False, regex=True)
]

# Clean the date by removing time zone and filtering out July, removing year
def clean_date(date_str):
    cleaned_str = re.sub(r' [A-Z]{3} ', ' ', date_str)
    return datetime.strptime(cleaned_str, "%a %b %d %H:%M:%S %Y")

tweets_college['date'] = tweets_college['date'].apply(clean_date)
tweets_college = tweets_college[tweets_college['date'].dt.month != 7]

# ensures in order by date
tweets_college = tweets_college.sort_values('date')


# Keep necessary columns and rename the target column
tweets_college = tweets_college[['target', 'date', 'text']]
tweets_college.rename(columns = {'target':'sentiment'}, inplace= True)

# Drop rows with missing values
tweets_college = tweets_college.dropna()

# Display cleaned dataset
print(tweets_college.shape)
tweets_college.head()

```

After cleaning, our dataset has 195 unique values. This is just right to do data analysis. Yet, not enough for modeling, but we shall focus on looking at trends of the data.

## Second Analysis - Hypothesis Testing

Despite looking at trends, we should run a hypothesis test to ensure our hypotheses hold true. Remember our hypotheses: 

- Null - H0 = There is no effect/relationship between anxious tweets and high-stress periods. 
- Alternate - H1 = There is a strong relationship between anxious tweets and high-stress periods in undergraduate students.
- Test Statistic = Mean of sentiment in high stress period and nonstress period


Performing the test, we are using t-test to compare the means of two groups:
- Group 1: High stress period (from June 7 to June 25, the typical final exam period for students, looking at the quarter system)
- Group 2: Nonstress period (any other data that isn't final week)

```python
# Define high-stress periods (for example, exam periods)
exam_start_date = pd.to_datetime('2009-06-07')
exam_end_date = pd.to_datetime('2009-06-25')

# Add a column to classify tweets as during high-stress period or not
tweets_college['high_stress_period'] =  tweets_college['date'].apply(lambda x: 1 if exam_start_date <= x <= exam_end_date else 0)

# Separate the data based on high-stress period classification
high_stress_sentiment = tweets_college[tweets_college['high_stress_period'] == 1]['sentiment']
non_stress_sentiment = tweets_college[tweets_college['high_stress_period'] == 0]['sentiment']

# Perform a T-test to compare the means of the two groups
t_stat, p_value = ttest_ind(high_stress_sentiment, non_stress_sentiment)

# Print the result
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in sentiment between high-stress and non-high-stress periods.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

Looking at the results, we see that we reject the null hypothesis. There is signficant evidence that implies that strong relationship between high stress periods and anxious tweets with undergraduate students. We assume that is because the distribution flattens togethers the end towards final week, implying only negative sentiment during final week.
